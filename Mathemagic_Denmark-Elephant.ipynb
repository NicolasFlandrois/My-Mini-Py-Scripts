# Mathemagic

Choose a random number (integer): a

To that number multiply by 3.
To this result add 4.
To that result multiply by 3, once more.
And finally add 1 to this result.

Then sum up each digits together from that number.
Repeat and do so until the result has only 1 digit.

This result, convert into a Letter.
(A = 1, B = 2, C = 3, ...etc..., I = 9)

Great, from that given letter choose a (european) country starting by this letter. (Write it down)

Take the last result from which you converted into a letter, and multiply it by 4.
From that result add 7.

Then sum up each digits together from that number.
Repeat and do so until the result has only 1 digit.

This result, convert into a Letter.
(A = 1, B = 2, C = 3, ...etc..., I = 9)

Great, from that given letter choose an animal (Mammal Herbivore) starting by this letter. (Write it down)

## Create a Script for this mathemagic

```python
from random import randint
```

```python
def looping(a:int):
    alpha_dict = {1:'A', 2:'B', 3:'C', 4:'D', 5:'E', 6:'F', 7:'G', 8:'H', 9:'I'}
    while True:
        b = [int(n) for n in str(a)]
        c = sum(b)
        a = c
        if len(b) <= 1:
            break
    return c, alpha_dict[c]
```

```python
def methamagic(a:int):
    print("Initial Integer Random Choice:\t", a)
    tmp = looping((((a*3)+4)*3)+1)
    print("Country Letter:\t", tmp[1])
    print("Animal Letter:\t", looping((tmp[0]*4)+7)[1])
```

```python
methamagic(a)
```
